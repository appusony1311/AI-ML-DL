
Pair plots

When you generalize joint plots to datasets of larger dimensions, you end up with pair plots. This is very useful for exploring correlations between multidimensional data, when you'd like to plot all pairs of values against each other.

We'll demo this with the well-known Iris dataset, which lists measurements of petals and sepals of three iris species:


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
print(iris.head())
print(iris.describe())
print(iris.columns)
print(iris.shape)
```

```python
sns.set_style("whitegrid");
sns.pairplot(iris, hue="species", size=3);
plt.show()
```

```python
ax = sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
ax = sns.swarmplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
ax = sns.swarmplot(x='species', y='petal_width', data=iris)
plt.show()
```

```python
sns.boxplot(x='species',y='petal_length', data=iris)
plt.show()
```

```python
sns.violinplot(x='species',y='petal_length', data=iris)
plt.show()
```

```python
g = sns.FacetGrid(iris, hue="species", size=5) 
g.map(sns.distplot, "petal_width").add_legend() 
plt.show()
```

```python
g = sns.FacetGrid(iris, hue="species", size=5) 
g.map(sns.distplot, "petal_length").add_legend() 
plt.show()
```

```python
g = sns.FacetGrid(iris, hue="species", size=5) 
g.map(sns.distplot, "sepal_length").add_legend() 
plt.show()
```

```python
g = sns.FacetGrid(iris, hue="species", size=5) 
g.map(sns.distplot, "sepal_width").add_legend() 
plt.show()
```
